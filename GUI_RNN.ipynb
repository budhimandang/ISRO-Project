```python
from tkinter import *
from tkinter import filedialog
from tkinter import ttk, StringVar, IntVar
from PIL import ImageTk, Image
from osgeo import gdal
from tkinter import messagebox
from keras.layers import *
import matplotlib.pyplot as plt
import os
import numpy as np
import tkinter.messagebox
import keras
from keras import Sequential
from keras.utils import np_utils
from PIL import Image
import string
import matplotlib
import rnncode
import array
import sys
import time
from threading import Thread
import io
from contextlib import redirect_stdout


class NEURAL:
    def __init__(self, master):

        self.textfile = StringVar()
        self.texthdr = StringVar()
        self.textweight = StringVar()
        self.checkbox_train = IntVar()
        self.checkbox_test = IntVar()
        self.yaxis=IntVar()
        self.cls=IntVar()
        self.data=[]
        self.layer = []
        self.value=0
        self.datacount=0
        self.flag=0
        self.layercount=0
        self.row=IntVar()
        self.col=IntVar()
        self.band=IntVar()
        self.dtypevalue=IntVar()
                
        self.master = master
        #self.master.configure(background='#80c1ff')
        self.master.geometry('780x620')
        self.master.resizable(0, 0)
        
        tkinter.ttk.Style().configure('s.TCheckbutton', font=('times new roman', 10, 'bold'), foreground='black',
                                      background='light grey')
        separator1 = Frame(self.master, height=7, background='grey20')
        separator1.pack(fill=X)
        separator2 = Frame(self.master, width=7, background='grey20')
        separator2.pack(fill=Y, side=LEFT)
        separator3 = Frame(self.master, height=7, background='grey20')
        separator3.pack(fill=X, side=BOTTOM)
        separator4 = Frame(self.master, width=7, bd=10, background='grey20')
        separator4.pack(fill=Y, side=RIGHT)
        self.tab_control = ttk.Notebook(self.master)
        self.tab2 = Frame(self.tab_control)
        self.tab3 = Frame(self.tab_control,background='grey20')
        self.tab_control.add(self.tab2, text='  Training and Testing  ')
        self.tab_control.add(self.tab3, text='  Load Training Data  ')
        self.tab_control.tab(1, state="disabled")

        self.tab_control2 = ttk.Notebook(self.tab2)
        # self.tab11 = ttk.Frame(self.tab_control2)
        self.tab11=Frame(self.tab_control2,height=300,background='gray20')
        self.tab_control2.add(self.tab11, text='Recurrent Neural Network')
        self.tab_control2.pack(expand=1, fill='both')
        self.tab_control.pack(expand=1, fill='both')
        # tkinter.ttk.Style().configure('TNotebook.Tab', foreground='black',background=mainWcolor)
        
        self.putrnn()

    def putrnn(self):
        self.textweight.set("")
        self.bt_image = Button(self.tab11, text='*Choose Image', foreground='black', background='light grey',command=self.img)
        self.bt_image.place(relx=0.10, rely=0.08, relwidth=0.125, relheight=0.05)
        self.lbl_image = Label(self.tab11, textvariable=self.textfile, relief="sunken", width=50)
        self.lbl_image.place(relx=0.30, rely=0.08, relwidth=0.48, relheight=0.04)

        self.bt_header = Button(self.tab11, text='*Choose Header', foreground='black', background='light grey',command=self.header)
        self.bt_header.place(relx=0.10, rely=0.18, relwidth=0.125, relheight=0.05)
        self.lbl_header = Label(self.tab11, textvariable=self.texthdr, relief="sunken", width=50)
        self.lbl_header.place(relx=0.30, rely=0.18, relwidth=0.48, relheight=0.04)

        back2 = Frame(self.tab11, background='white')
        back2.place(relheight=0.06, relwidth=0.11, relx=0.03, rely=0.28)
        self.rb1 = tkinter.ttk.Checkbutton(back2, text='Training', variable=self.checkbox_train, style='s.TCheckbutton')
        self.rb1.place(relx=0.065, rely=0.13, relwidth=0.87, relheight=0.7)

        separator5 = Frame(self.tab11, height=2, width=600, background='white')
        separator5.place(relx=0.18, rely=0.31, relwidth=0.75)

        self.label_activation = Label(self.tab11, text="*Select Activation function",background='grey20',foreground='white')
        self.label_activation.place(relx=0.01, rely=0.38, relwidth=0.3, relheight=0.05)

        self.combobox_activation = ttk.Combobox(self.tab11, values=[
                                                                    "relu",
                                                                    "sigmoid",
                                                                    "tanh"])
        self.combobox_activation.place(relx=0.3, rely=0.39, relwidth=0.15, relheight=0.04)
        self.combobox_activation.current(0)

        self.label_pooling = Label(self.tab11, text="*Select pooling",background='grey20',foreground='white')
        self.label_pooling.place(relx=0.01, rely=0.43, relwidth=0.3, relheight=0.05)

        self.combobox_pooling = ttk.Combobox(self.tab11, values=[
                                                                "Max Pooling",
                                                                "Average Pooling"])
        self.combobox_pooling.place(relx=0.3, rely=0.44, relwidth=0.15, relheight=0.04)
        self.combobox_pooling.current(0)

        self.label_threshold = Label(self.tab11, text="*Enter threshold",background='grey20',foreground='white')
        self.label_threshold.place(relx=0.01, rely=0.48, relwidth=0.3, relheight=0.05)

        self.entry_threshold = Entry(self.tab11)
        self.entry_threshold.place(relx=0.3, rely=0.49, relwidth=0.2, relheight=0.035)

        self.label_epoch = Label(self.tab11, text="*Enter Epoch",background='grey20',foreground='white')
        self.label_epoch.place(relx=0.01, rely=0.53, relwidth=0.3, relheight=0.05)

        self.entry_epoch = Entry(self.tab11)
        self.entry_epoch.place(relx=0.3, rely=0.54, relwidth=0.2, relheight=0.035)

        self.label_classes = Label(self.tab11, text="*Enter no of classes",background='grey20',foreground='white')
        self.label_classes.place(relx=0.01, rely=0.58, relwidth=0.3, relheight=0.05)

        self.entry_classes = Entry(self.tab11)
        self.entry_classes.place(relx=0.3, rely=0.59, relwidth=0.2, relheight=0.035)

        back3 = Frame(self.tab11, background='white')
        back3.place(relheight=0.06, relwidth=0.11, relx=0.03, rely=0.69)
        self.rb2 = tkinter.ttk.Checkbutton(back3, text='Testing', variable=self.checkbox_test, style='s.TCheckbutton')
        self.rb2.place(relx=0.065, rely=0.13, relwidth=0.87, relheight=0.7)
        separator6 = Frame(self.tab11, height=2, width=600, background='white')
        separator6.place(relx=0.18, rely=0.715, relwidth=0.75)

        self.bt_load = Button(self.tab11, text='Load Model', foreground='black', background='light grey', command=self.load_model)
        self.bt_load.place(relx=0.10, rely=0.79, relwidth=0.125, relheight=0.05)
        self.lbl_model = Label(self.tab11, textvariable=self.textweight, relief="sunken", width=50)
        self.lbl_model.place(relx=0.30, rely=0.79, relwidth=0.48, relheight=0.04)

        self.bt_next = Button(self.tab11, text='Next', foreground='black', background='light grey',command=self.get_training_data)
        self.bt_next.place(relx=0.43, rely=0.90, relwidth=0.12, relheight=0.05)

    def load_model(self):
        filename = filedialog.askopenfilename(initialdir="/", title="Select file", defaultextension='.txt')
        self.textweight.set(filename)

    def img(self):
        filename = filedialog.askopenfilename(initialdir="/", title="Select file",
                                              filetypes=(("all files", "*.*"), ("Text file", "*.txt*")))
        self.textfile.set(filename)
        self.image = filename
        print(self.image)

    def header(self):
        filename = filedialog.askopenfilename(initialdir="/", title="Select file",
                                              filetypes=(("Header files", "*.hdr*"), ("All files", "*.*")))
        self.texthdr.set(filename)
        self.head = filename
        print(self.head)

    def addfiles(self):
        if len(self.entry_classes.get()) == 0:
            messagebox.showerror("Error", "Enter number of training classes!")
        else:            
            add =  filedialog.askopenfilename(initialdir = "/",title = "Select file",filetypes = (("all files","*.*"),("Text file","*.txt")))
            r=self.entry_classes.get()
            self.cls=int(r)+1
            self.data.append(add)
            self.listbox1.config(state="normal")
            self.listbox1.delete(0,END)
            for ent in self.data:
                self.listbox1.insert(END, ent+'\n')
            self.listbox1.config(state="disabled")
            if len(self.data)>=(self.cls-1):
                self.btnadd.config(state="disabled")
            elif len(self.data)<(self.cls-1):
                self.btnadd.config(state="normal")
            self.datacount=self.datacount+1
            print(len(self.data))
            print(self.cls)
                        
    def deletefiles(self):
        self.data=self.data[:-1]
        self.listbox1.config(state="normal")
        self.listbox1.delete(END)
        self.listbox1.config(state="disabled")
        if len(self.data)<self.cls:
            self.btnadd.config(state="normal")
        self.datacount=self.datacount-1
                    
    def putloaddata(self):
        back4 = Frame(self.tab3, width=100)
        back4.pack(fill=X, padx=100, pady=20)
        self.btnadd = Button(self.tab3, text="Add Files", foreground='black', background='light grey',
                             command=self.addfiles)
        self.btnadd.pack(fill=X, padx=300, pady=10)

        self.btndelete = Button(self.tab3, text="Remove Files", foreground='black', background='light grey',
                                command=self.deletefiles)
        self.btndelete.pack(fill=X, padx=300, pady=10)
        separator5 = Frame(self.tab3, height=1, background='grey')
        separator5.pack(fill=X, padx=150)
        frm = Frame(self.tab3, width=100)
        frm.pack(fill=X, padx=100, pady=30)
        self.listbox1 = Listbox(frm, height=5, font=("Times new roman", 14))
        self.listbox1.grid(column=0, row=0, sticky=(N, W, E, S))
        s = ttk.Scrollbar(frm, orient=VERTICAL, command=self.listbox1.yview)
        s.grid(column=2, row=0, sticky=(N, S))
        self.listbox1['yscrollcommand'] = s.set

        frm.grid_columnconfigure(0, weight=1)
        frm.grid_rowconfigure(0, weight=1)
        self.btn_next1 = Button(self.tab3, text="Next", foreground='black', background='light grey',
                             command=self.getlayers)
        self.btn_next1.place(x=450, y=400, width=100)


    def getlayers(self):
        r=int(self.entry_classes.get())
        if r!=self.datacount:
            messagebox.showerror("Error! Add class")
        else:
            self.window7=ttk.tkinter.Toplevel()
            self.window7.geometry("500x650")
            self.scrollbar1 = Scrollbar(self.window7)
            self.scrollbar1.pack(side=RIGHT, fill=Y)

            self.T2 = Listbox(self.window7, yscrollcommand=self.scrollbar1.set)
            self.l11 = Label(self.window7, text="Add Layers sequentially in the Neural Network")
            self.l11.place(x=50, y=50, width=350)
            self.l22 = Label(self.window7, text="Convolution Layer")
            self.l22.place(x=20, y=100, width=100)
            pool=str(self.combobox_pooling.get())
            poolc="Max Pooling"
            if(pool==poolc):
                self.value=1
            else:
                self.value=2
            if(self.value is 1):
                self.l33 = Label(self.window7, text="MaxPooling Layer")
                self.l33.place(x=20, y=150, width=120)
            else:
                self.l33 = Label(self.window7, text="AveragePooling Layer")
                self.l33.place(x=20, y=150, width=120)
            self.l34 = Label(self.window7, text="LSTM Layer")
            self.l34.place(x=20, y=200, width=100)

            self.b22 = ttk.Button(self.window7, text='+', cursor="plus", style='add.TButton', command=self.addConv)
            self.b22.place(x=250, y=100, width=25)
            self.b44 = ttk.Button(self.window7, text='+', cursor="plus", style='add.TButton', command=self.addPool)
            self.b44.place(x=250, y=150, width=25)
            self.b55 = ttk.Button(self.window7, text='+', cursor="plus", style='add.TButton', command=self.addLSTM)
            self.b55.place(x=250, y=200, width=25)
            self.b33 = ttk.Button(self.window7, text='-', cursor="plus", style='remove.TButton', command=self.delete)
            self.b33.place(x=250, y=250, width=25)

            self.T2.place(x=75, y=350, width=150)
            self.b66 = ttk.Button(self.window7, text='Classify', command=self.loadrnn)
            self.b66.place(x=75, y=300, width=100)
            self.scrollbar1.config(command=self.T2.yview)
        
    def addConv(self):
        self.layer.append("Convolution")
        self.T2.config(state="normal")
        self.T2.delete('0', END)
        for ent in self.layer:
            self.T2.insert(END, ent + '\n')
        self.T2.config(state="disabled")
        self.layercount=self.layercount+1

    def delete(self):
        self.layer = self.layer[:-1]
        self.T2.config(state="normal")
        self.T2.delete('0', END)
        for ent in self.layer:
            self.T2.insert(END, ent + '\n')
        self.T2.config(state="disabled")
        self.layercount=self.layercount-1

    def addPool(self):
        if(self.value is 1):
            self.layer.append("MaxPooling")
        else:
            self.layer.append("AveragePooling")
        self.T2.config(state="normal")
        self.T2.delete('0', END)
        for ent in self.layer:
            self.T2.insert(END, ent + '\n')
        self.T2.config(state="disabled")
        self.layercount=self.layercount+1

    def addLSTM(self):
        self.layer.append("LSTM")
        self.T2.config(state="normal")
        self.T2.delete('0', END)
        for ent in self.layer:
            self.T2.insert(END, ent + '\n')
        self.T2.config(state="disabled")
        self.layercount=self.layercount+1

    def save_values(self):
        w=self.entry_epoch.get()
        print(w)
        r=self.entry_classes.get()
        print(r)
        t=self.entry_threshold.get()
        act=self.combobox_activation.get()
        pool=str(self.combobox_pooling.get())
        poolc="MaxPooling"
        if(pool==poolc):
            self.value=1
        else:
            self.value=2
        print(self.value)
        value=str(self.value)
        self.file = open('saved_val.txt','w+')
        end='//null//'
        self.file.write('epoch:'+w+end)
        self.file.write('cls:'+r+end)
        self.file.write('threshold:'+t+end)
        self.file.write('activation:'+act+end)
        self.file.write('pooling:'+pool+end)
        self.file.write('poolingvalue:'+value+end)
        i=self.textfile.get()
        self.file.write('imgpath;'+i+end)
        h=self.texthdr.get()
        self.file.write('headerpath:'+h+end) 
        self.row,self.col,self.band,self.dtypevalue=self.headerreader(self.head)
        self.file.write('row:'+str(self.row)+end)
        self.file.write('col:'+str(self.col)+end)
        self.file.write('band:'+str(self.band)+end)
        self.file.write('dtype:'+str(self.dtypevalue)+end)
        self.file.close()
        self.flag=1
        
    def loadrnn(self):
        self.window7.destroy()
        self.tab_control.tab(0,state="normal")
        self.tab_control.tab(1,state="disabled")
        if(self.checkbox_train.get() == 1 and self.checkbox_test.get() == 1):
            self.save_values()
            rnncode.testtrainmodel(self.data,self.layer)
        elif(self.checkbox_train.get() == 1):
            self.save_values()
            rnncode.trainmodel(self.data,self.layer)        

    
    def headerreader(self,headpath):
        dtypevalue=self.getdvalue(headpath)
        i=1
        r=-1
        c=-1
        b=-1
        pat1 = re.compile(r"\brows\b", re.IGNORECASE) # but matches because pattern ignores case
        pat2 = re.compile(r"\bcols\b", re.IGNORECASE) # but matches because pattern ignores case
        pat3 = re.compile(r"\bbands\b", re.IGNORECASE)
        pat4 = re.compile(r"\bbildatatype\b", re.IGNORECASE)# but matches because pattern ignores case
        lines = [] #Declare an empty list named "lines"
        with open (headpath, 'rt') as in_file: # Open file txt for reading of text data.
            for line in in_file:
                if pat1.search(line) != None:
                    lines.append(line) 
                    i=i+1
                    r=line.split(" ",1)[1] 
                    print("No. of rows:")
                    print(r)

                elif pat2.search(line) != None:
                    lines.append(line) 
                    i=i+1
                    c=line.split(" ",1)[1] 
                    print("No. of columns:")
                    print(c)

                elif pat3.search(line) != None:
                    lines.append(line) 
                    i=i+1
                    b=line.split(" ",1)[1] 
                    print("Bands:")
                    print(b)
                else:
                    i=i+1

        if( r==-1 or b==-1 or c==-1):
            print("ERROR:wrong header file")

        in_file.close()
        rr=int(r)
        cc=int(c)
        bb=int(b)
        print("Datatype:")
        print(dtypevalue)
        
        return rr,cc,bb,dtypevalue
    
    def getdvalue(self,file):
        if file.endswith('.hdr'):
            f= open(file,"r")
        else:
            return None
        i=1
        for line in f.readlines():
            line = str(line.lower())
            line = line.strip().lower()
            if str(line) == "datatype: u16" or (str(line) == "datatype: s16"):
                datavalue=16
            if str(line) == "datatype: u8":
                datavalue=8
            i=i+1
        return datavalue

    
    def get_training_data(self):
        if (self.checkbox_test.get() != 1 and self.checkbox_train.get() != 1):
            messagebox.showerror("Error", "Select at least one of training or testing!")
        elif (self.checkbox_test.get() == 1 and self.checkbox_train.get() != 1):
            val1 = self.textfile.get().split("/")
            val2 = self.texthdr.get().split("/")
            if len(self.textfile.get()) == 0:
                messagebox.showerror("Error", "Image file field cannot be empty!")
            elif len(self.texthdr.get()) == 0:
                messagebox.showerror("Error", "Header file field cannot be empty!")
            elif ((val1[-1] + ".hdr") != (val2[-1])):
                messagebox.showerror("Error", "Wrong header file")
            elif len(self.textweight.get()) == 0:
                messagebox.showerror("Error", "Weights file needed!")
            else:
                rnncode.modeltest(self.textweight.get(),self.texthdr.get(),self.textfile.get())
        else:
            val1 = self.textfile.get().split("/")
            val2 = self.texthdr.get().split("/")
            if len(self.entry_epoch.get()) == 0:
                messagebox.showerror("Error", "Epochs field cannt be empty!")
            elif len(self.entry_classes.get()) == 0:
                messagebox.showerror("Error", "Training classes field cannot be empty!")
            elif len(self.entry_threshold.get()) == 0:
                messagebox.showerror("Error", "Threshold field cannot be empty!")
            elif len(self.textfile.get()) == 0:
                messagebox.showerror("Error", "Choose Image file first!")
            elif len(self.texthdr.get()) == 0:
                messagebox.showerror("Error", "Choose header file first!")
            elif ((val1[-1] + ".hdr") != (val2[-1])):
                messagebox.showerror("Error", "Wrong header file!")
            else:
                if (self.checkbox_test.get() == 1 and self.checkbox_train.get() == 1):
                    if len(self.textweight.get()) == 0:
                        messagebox.showerror("Error", "Weights file needed!")
                    else:
                        self.tab_control.tab(1, state="normal")
                        self.tab_control.tab(0, state="disabled")
                        self.putloaddata()
                        self.rb1.config(state="disabled")
                else:
                    self.tab_control.tab(1, state="normal")
                    self.tab_control.tab(0, state="disabled")
                    self.putloaddata()
                    self.rb1.config(state="disabled")


if __name__ == '__main__':
    root = Tk()
    root.title('Image Classification using rnn')
    obj = NEURAL(root)
    root.mainloop()

```
